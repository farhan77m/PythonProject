```python
# Check for missing values in num_lab_procedures
missing_values = df['num_lab_procedures'].isnull().sum()
print(f"\nNumber of missing values in num_lab_procedures: {missing_values}")

# Replace missing values with the mean
df['num_lab_procedures'].fillna(df['num_lab_procedures'].mean(), inplace=True)
```

```python
# Check for missing values in time_in_hospital
missing_values = df['time_in_hospital'].isnull().sum()
print(f"Number of missing values in time_in_hospital: {missing_values}")

# Replace missing values with the median
df['time_in_hospital'].fillna(df['time_in_hospital'].median(), inplace=True)
```

```python
# Vizualization for time_in_hospital by gender
plt.figure(figsize=(12, 8))
sns.boxplot(x='gender', y='time_in_hospital', data=df)
plt.title('Distribution of Time in Hospital by Gender')
plt.xlabel('Gender')
plt.ylabel('Time in Hospital')
plt.show()
```

```python
# Visualization for Number of Lab Procedure by Admmission Type Id
plt.figure(figsize=(12, 8))
sns.boxplot(x='admission_type_id', y='num_lab_procedures', data=df)
plt.title('Distribution of Number of Lab Procedures by Admission Type Id')
plt.xlabel('Admission Type Id')
plt.ylabel('Number of Lab Procedures')
plt.show()
```
